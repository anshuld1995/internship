# CAR PRICE PREDICTION
With the covid 19 impact in the market, we have seen lot of changes in the car market. Now some cars are in demand hence making them costly and some are not in demand hence cheaper. One of our clients works with small traders, who sell used cars. With the change in market due to covid 19 impact, our client is facing problems with their previous car price valuation machine learning models. So, they are looking for new machine learning models from new data. We have to make car price valuation model. This project contains two phase

Data Collection Phase You have to scrape at least 5000 used cars data. You can scrape more data as well, it’s up to you. more the data better the model

In this section You need to scrape the data of used cars from websites (Olx, cardekho, Cars24 etc.) You need web scraping for this. You have to fetch data for different locations. The number of columns for data doesn’t have limit, it’s up to you and your creativity. Generally, these columns are Brand, model, variant, manufacturing year, driven kilometers, fuel, number of owners, location and at last target variable Price of the car. This data is to give you a hint about important variables in used car model. You can make changes to it, you can add or you can remove some columns, it completely depends on the website from which you are fetching the data.

```python
import pandas as pd
import selenium
from selenium import webdriver
from time import sleep
import time 
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException,ElementNotInteractableException
import warnings 
warnings.filterwarnings('ignore')
```

# Web_Scrapping the Used Car details from "cardekho.com"¶

```python
driver = webdriver.Chrome(r"E:\New folder\chromedriver.exe") 
```

```python
url = 'https://www.cardekho.com/' 
driver.get(url)
```

```python
time.sleep(4)

#scrolling the cursor below in order to get number of details added on the page
for _ in range(50):
    driver.execute_script("window.scrollBy(0,200)")
    
time.sleep(4)
    
    
city_links=[]
#links of each car data
cities = driver.find_elements_by_xpath('//ul[@class="listing gsc_col-md-9 gsc_col-lg-9"]/li/a')

#saving the links in a list which are related to each location of available cars
for i in cities:
    try: 
        city_links.append(i.get_attribute('href'))
    except NoSuchElementException:
        continue
```

```python
city_links
```

# Let's scrap the car details from each location page

# Ahmedabad

```python
#To extract the Car details from the page with "Ahmedabad" Location

driver.get(city_links[0])
    
links1=[]
    
time.sleep(12)

#scrolling the cursor below in order to get number of details added on the page
for _ in range(2000):
    driver.execute_script("window.scrollBy(0,500)")

    
    
for i in driver.find_elements_by_xpath('//div[@class="gsc_col-xs-7  carsName"]/a'):
    try: 
        links1.append(i.get_attribute('href'))
    except NoSuchElementException:
        continue
```

```python
links1
```

```python
brand=[]
model=[]
variant=[]
Driven_KM=[]
fuel=[]
price=[]



time.sleep(8)
start=0
end=4
for page in range(start,end):
    brands=driver.find_elements_by_xpath('//div[@class="gsc_col-xs-7  carsName"]/a')
    for i in brands:
        brand.append(i.text)
        
    prices=driver.find_elements_by_xpath('//span[@class="amnt "]')
    for i in prices:
        price.append(i.text)
        
    models=driver.find_elements_by_xpath('//div[@class="car_variant"]')
    for i in models:
        model.append(i.text)  
        
    kms=driver.find_elements_by_xpath(('//div[@class="truncate dotlist-2"]/div[1]'))
    for i in kms:
        Driven_KM.append(i.text)
        
    fuels=driver.find_elements_by_xpath(('//div[@class="truncate dotlist-2"]/div[2]'))
    for i in fuels:
        fuel.append(i.text)
        
    variants=driver.find_elements_by_xpath(('//div[@class="truncate dotlist-2"]/div[3]'))
    for i in variants:
        variant.append(i.text)
        
    
```

```python
            
Ahmedabad=pd.DataFrame({
'Brand':brand[:],                 
'Price':price[:],
'Model':model[:],
'KMS_driven':Driven_KM[:],
'Fuel':fuel[:],
'Variant':variant[:]})
Ahmedabad
```

```python
B = {}
B['Location'] = []

for _ in range(848):
    B['Location'].append('Ahmedabad')
B = pd.DataFrame(B)




Ahmedabad['Location'] = B

Ahmedabad
```

```python
Ahmedabad.to_csv('Ahmedabad_CarDetails.csv')
```
