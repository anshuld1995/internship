# 1) Write a python program to display all the header tags from wikipedia.org.


```python
def header(url):
    page = requests.get(url)
    print(page)
    soup_wiki = BeautifulSoup(page.content)
    titles = []
    for i in soup_wiki.find_all(['h1','h2','h3','h4','h5','h6']):
        titles.append(i.text.replace('\n',''))
    print("List all the Header Tags",*titles,sep='\n')    
    
```

```python
header('https://en.wikipedia.org/wiki/Main_Page')
```

# 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)and make data frame



```python
def Movies(url1,url2):
    imdb_page1 = requests.get(url1)
    imdb_page2 = requests.get(url2)
    
    print(imdb_page1,imdb_page2)
    
    soup_imdb1 = BeautifulSoup(imdb_page1.content)
    soup_imdb2 = BeautifulSoup(imdb_page2.content)
    
    movies = []
    for i in soup_imdb1.find_all('h3','lister-item-header'):
        movies.append(i.text.replace('\n','').strip('()').split("("))
    
    for j in soup_imdb2.find_all('h3','lister-item-header'):
        movies.append(j.text.replace('\n','').strip("()").split("("))
    
    name=[]
    release_year = []
    for k in range(len(movies)):
        name.append(movies[k][0])
        release_year.append(movies[k][1])
    
    movie_name =[]
    for i in name:
        movie_name.append(i.split('.')[1])
    
    rat=[]
    for l in soup_imdb1.find_all('div',class_= 'ratings-bar'):
        rat.append(l.text.split()[0])
    for l in soup_imdb2.find_all('div',class_= 'ratings-bar'):
        rat.append(l.text.split()[0])
    
    import pandas as pd
    dict = {'Name':movie_name,'Rating': rat,'Year of Release': release_year }
    df = pd.DataFrame(dict)
    return(df)
Movies('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc','https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,dec')
```

# 3) Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame.


```python
def Indian_Movies(url1):
    imbd_page1=requests.get(url1)
    soup_imbd1 = BeautifulSoup(imbd_page1.content)
       
    movies = []
    for i in soup_imbd1.find_all('h3',class_ ='lister-item-header'):
        movies.append(i.text.replace('\n','').strip('()').split("("))
            
    name=[]
    release_year = []
    for k in range(len(movies)):
        name.append(movies[k][0])
        release_year.append(movies[k][1])
    
    movie_name =[]
    for i in name:
        movie_name.append(i.split('.')[1])
    
    rat=[]
    for l in soup_imbd1.find_all('div',class_= 'ipl-rating-widget'):
        rat.append(l.text.split()[0])
    
    import pandas as pd
    dict = {'Name':movie_name,'Rating': rat,'Year of Release': release_year }
    df = pd.DataFrame(dict)
    return(df)
```

# 5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
b) Top 10 ODI Batsmen along with the records of their team and rating.
c) Top 10 ODI bowlers along with the records of their team and rating.# 

```python
# a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
def ODI_RANKING_MENS(url):
    ODI_teams = requests.get(url)
    ODI = BeautifulSoup(ODI_teams.content)
    team_name = []
    for i in ODI.find_all('span',class_ ='u-hide-phablet'):
        team_name.append(i.text.replace("\n"," "))
        team_name = team_name[:10]
    records = []
    for i in ODI.find_all('tr',class_ ='rankings-block__banner'):
        records.append(i.text.replace("\n"," ").split()[4:])
    for j in ODI.find_all('tr',class_ ='table-body'):
        records.append(j.text.replace("\n"," ").split()[-3:])
        records =records[:10]
    matches = []
    points =[]
    rating =[]
    for k in range(len(records)):
        matches.append(records[k][0])
        points.append(records[k][1])
        rating.append(records[k][2])
    import pandas as pd
    df = pd.DataFrame({'Team':team_name,'Matches':matches,'Points':points,'Rating':rating})
    return(df)

ODI_RANKING_MENS('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

```python
#b) Top 10 ODI Batsmen along with the records of their team and rating.

def ODI_top10_bats(url):
    Batsman = requests.get(url)
    print(Batsman)
    ODI_batsman = BeautifulSoup(Batsman.content)
    
    Batsman_name = []
    for i in ODI_batsman.find_all('div',class_ ='rankings-block__banner--name-large'):
        Batsman_name.append(i.text.replace("\n"," "))
    
    for j in ODI_batsman.find_all('td',class_ ='table-body__cell rankings-table__name name'):
        Batsman_name.append(j.text.replace("\n"," "))
        Batsman_name = Batsman_name[:10]
        
    team= []
    for i in ODI_batsman.find_all('div',class_ ='rankings-block__banner--nationality'):
        team.append(i.text.replace("\n",""))
    for j in ODI_batsman.find_all('span',class_ ='table-body__logo-text'):
        team.append(j.text.replace("\n"," "))
        team=team[:10]
        
    Rating=[]
    for i in ODI_batsman.find_all('div',class_ ='rankings-block__banner--rating'):
        Rating.append(i.text.replace("\n",""))
    for j in ODI_batsman.find_all('td',class_ ='table-body__cell rating'):
        Rating.append(j.text.replace("\n"," "))
        Rating=Rating[:10]
        
    dict = {'Batsman name':Batsman_name,'Team':team,'Rating':Rating}
    import pandas as pd
    df = pd.DataFrame(dict)
    return(df)

ODI_top10_bats('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting')
```

```python
#(C) Top 10 ODI bowlers along with the records of their team and rating.

def ODI_top10_bowler(url):
    Bowler = requests.get(url)
    print(Bowler)
    ODI_bowler = BeautifulSoup(Bowler.content)
    
    Bowler_name = []
    for i in ODI_bowler.find_all('div',class_ ='rankings-block__banner--name-large'):
        Bowler_name.append(i.text.replace("\n"," "))
    
    for j in ODI_bowler.find_all('td',class_ ='table-body__cell rankings-table__name name'):
        Bowler_name.append(j.text.replace("\n",""))
        Bowler_name = Bowler_name[:10]
        
    team= []
    for i in ODI_bowler.find_all('div',class_ ='rankings-block__banner--nationality'):
        team.append(i.text.replace("\n",""))
    for j in ODI_bowler.find_all('span',class_ ='table-body__logo-text'):
        team.append(j.text.replace("\n"," "))
        team=team[:10]
        
    Rating=[]
    for i in ODI_bowler.find_all('div',class_ ='rankings-block__banner--rating'):
        Rating.append(i.text.replace("\n",""))
    for j in ODI_bowler.find_all('td',class_ ='table-body__cell rating'):
        Rating.append(j.text.replace("\n"," "))
        Rating=Rating[:10]
        
    dict = {'Bowler name':Bowler_name,'Team':team,'Rating':Rating}
    import pandas as pd
    df = pd.DataFrame(dict)
    return(df)
```

# Write a python program to scrape mentioned details from dineout.co.in 

# i) Restaurant name
ii) Cuisine
iii) Location 
iv) Ratings
v) Image URL

```python
def dineout(url):
    dineout = requests.get(url)
    print(dineout)
    dineout_soup = BeautifulSoup(dineout.content)
    
    restaurant_name = []
    for i in dineout_soup.find_all('div',class_ ='restnt-info cursor'):
        restaurant_name.append(i.text)
        
    cuisine = []
    for i in dineout_soup.find_all('span',class_ ='double-line-ellipsis'):
        cuisine.append(i.text.replace('₹','').split('|')[1])
    
    location=[]
    for i in dineout_soup.find_all('div',class_ ='restnt-loc ellipsis'):
        location.append(i.text)
        
    rating = []
    for i in dineout_soup.find_all('div',class_='restnt-rating rating-4'):
        rating.append(i.text)

    image_url = []
    for i in dineout_soup.find_all('img',class_='no-img'):
        image_url.append(i['data-src'])
    
    import pandas as pd
    dic = {'Restaurant name':restaurant_name,'Cuisine':cuisine,'Location':location,'Rating':rating,'image_url':image_url}
    df = pd.DataFrame(dic)
    return(df)
dineout('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```
